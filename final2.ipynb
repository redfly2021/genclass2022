```python
from glob import glob
import numpy as np
from PIL import Image

def pngs(root):
    return sorted(glob(f'{root}/*.png'))

def loads(paths, resize=1024):
    def load_(p, sz=resize):
        img = Image.open(p).convert('RGB').resize((sz,sz))
        return np.array(img)
    ims = np.concatenate(list(map(load_, paths)), axis=1)
    return Image.fromarray(ims)
```

```python

# 이미지를 Data를 Imageio로 동영상으로 변환
def generate_mp4(out_name, images, kwargs):
    writer = imageio.get_writer(out_name , **kwargs)
    for image in images:
        writer.append_data(image)
    writer.close()

generate_mp4('remastered.mp4', frames, {'fps':24})
```
